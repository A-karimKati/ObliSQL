```python
import pandas as pd
import seaborn as sns
import os
```

```python

def read_text_files(directory,component,number,label):
    text_data = []
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            with open(os.path.join(directory, filename), 'r') as file:
                d={
                'In-Flight':0,
                'Throughput':0,
                'Latency(ms)': 0,
                'ScaleFactor': component,
                "# of Components":number,
                "Label":label
                }
                for line in file:
                    if line.startswith("In-Flight Requests"):
                        inFlight = int(line.split(' ')[2].strip())
                        d['In-Flight']=inFlight
                    if line.startswith("Ops/s"):
                        parts = line.split(',')
                        throughput = int(parts[1])
                        d['Throughput']=throughput/30
                    
                    if line.startswith("Average Latency"):
                        latency = int(line.split(' ')[2])
                        d['Latency(ms)']=latency
                
                text_data.append(d)
    return text_data
```

```python
scaleOne10ms = '/home/haseeb/Desktop/ReviewGraphs/Latency/Waffle/10MS/Scale1'
scaleTwo10ms = '/home/haseeb/Desktop/ReviewGraphs/Latency/Waffle/10MS/Scale2'
scaleThree10ms = '/home/haseeb/Desktop/ReviewGraphs/Latency/Waffle/10MS/Scale3'

scaleOne20ms = '/home/haseeb/Desktop/ReviewGraphs/Latency/Waffle/20MS/Scale1'
scaleTwo20ms = '/home/haseeb/Desktop/ReviewGraphs/Latency/Waffle/20MS/Scale2'
scaleThree20ms = '/home/haseeb/Desktop/ReviewGraphs/Latency/Waffle/20MS/Scale3'


scaleOne30ms = '/home/haseeb/Desktop/ReviewGraphs/Latency/Waffle/30MS/Scale1'
scaleTwo30ms = '/home/haseeb/Desktop/ReviewGraphs/Latency/Waffle/30MS/Scale2'
scaleThree30ms = '/home/haseeb/Desktop/ReviewGraphs/Latency/Waffle/30MS/Scale3'


readscaleOne10ms= read_text_files(scaleOne10ms, '1', 3, "10ms")
readscaleTwo10ms = read_text_files(scaleTwo10ms, '2', 3, "10ms")
readscaleThree10ms = read_text_files(scaleThree10ms, '3', 3, "10ms")


readscaleOne20ms= read_text_files(scaleOne20ms, '1', 3, "20ms")
readscaleTwo20ms = read_text_files(scaleTwo20ms, '2', 3, "20ms")
readscaleThree20ms = read_text_files(scaleThree20ms, '3', 3, "20ms")

readscaleOne30ms = read_text_files(scaleOne30ms, '1', 3, "30ms")
readscaleTwo30ms = read_text_files(scaleTwo30ms, '2', 3, "30ms")
readscaleThree30ms = read_text_files(scaleThree30ms, '3', 3, "30ms")

oneDF = pd.DataFrame(readscaleOne10ms)
twoDF = pd.DataFrame(readscaleTwo10ms)
threeDF = pd.DataFrame(readscaleThree10ms)
fourDF = pd.DataFrame(readscaleOne20ms)
fiveDF = pd.DataFrame(readscaleTwo20ms)
sixDF = pd.DataFrame(readscaleThree20ms)
sevenDF = pd.DataFrame(readscaleOne30ms)
eightDF = pd.DataFrame(readscaleTwo30ms)
nineDF = pd.DataFrame(readscaleThree30ms)

proxyDF = pd.concat([oneDF, twoDF, threeDF, fourDF, fiveDF, sixDF, sevenDF, eightDF, nineDF])
proxyDF.rename(columns={'ScaleFactor': 'Machines', 'Label': "Latency"}, inplace=True)
proxyDF

proxyDF.groupby(['Machines', 'Latency']).agg({
    'Throughput': ['mean'],
    'Latency(ms)': ['mean']
})
```
